```python
import sounddevice as sd
import soundfile as sf
import speech_recognition as sr
import pyttsx3
from groq import Groq
```

```python
from groq import Groq

client = Groq(
    api_key="PLACEHOLDER"
)
```

```python
engine = pyttsx3.init()

def speak(text):
    print("Assistant:", text)
    engine.say(text)
    engine.runAndWait()
```

```python
speak("Smart Santhe assistant started")
```

```python
def record_audio():

    print("🎤 Recording for 5 seconds...")

    duration = 5
    samplerate = 44100

    recording = sd.rec(int(duration * samplerate),
                       samplerate=samplerate,
                       channels=1)

    sd.wait()

    sf.write("voice.wav", recording, samplerate)

    print("Recording complete")
```

```python
def voice_to_text():

    r = sr.Recognizer()

    with sr.AudioFile("voice.wav") as source:
        audio = r.record(source)

    try:
        text = r.recognize_google(audio)
        print("You:", text)
        return text

    except:
        print("Could not understand")
        return None
```

```python
def ask_groq(message):

    completion = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {
                "role": "system",
                "content": "You are a helpful farming assistant for Smart Digital Santhe."
            },
            {
                "role": "user",
                "content": message
            }
        ],
        temperature=0.7,
        max_tokens=200
    )

    return completion.choices[0].message.content
```

```python
try:
    while True:

        record_audio()
        text = voice_to_text()

        if text is None:
            continue

        text = text.lower()

        # Voice stop command
        if text in ["stop", "exit", "quit"]:
            speak("Assistant stopped")
            break

        response = ask_groq(text)
        speak(response)

except KeyboardInterrupt:
    print("Voice assistant stopped manually")
```
